## Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:

## a)1/ 125∗x2+1,x0=0
## b) arctan(x),x0=1

```python
import sympy as sp

# Serie de Taylor
# 1. Función: f(x) = 1 / (25x^2 + 1), x0 = 0
x = sp.Symbol('x')
f1 = 1 / (25 * x**2 + 1)
x0_1 = 0

# Serie de Taylor para f1
n_terms = 4  # Número de términos de la serie
series_f1 = sp.series(f1, x, x0_1, n=n_terms)

print("Serie de Taylor para f(x) = 1 / (25x^2 + 1) alrededor de x0 = 0:")
print(series_f1)

# 2. Función: f(x) = arctan(x), x0 = 1
f2 = sp.atan(x)
x0_2 = 1

# Serie de Taylor para f2
series_f2 = sp.series(f2, x, x0_2, n=n_terms)

print("\nSerie de Taylor para f(x) = arctan(x) alrededor de x0 = 1:")
print(series_f2)

# Polinomio de Lagrange
# Datos para la interpolación de Lagrange
x_points = [0, 0.5, 1]
y_points_f1 = [f1.subs(x, xi) for xi in x_points]
y_points_f2 = [f2.subs(x, xi) for xi in x_points]

# Construcción del polinomio de Lagrange
def lagrange_polynomial(x_points, y_points):
    n = len(x_points)
    P = 0
    for i in range(n):
        L = 1
        for j in range(n):
            if i != j:
                L *= (x - x_points[j]) / (x_points[i] - x_points[j])
        P += y_points[i] * L
    return sp.simplify(P)

lagrange_f1 = lagrange_polynomial(x_points, y_points_f1)
lagrange_f2 = lagrange_polynomial(x_points, y_points_f2)

print("\nPolinomio de Lagrange para f(x) = 1 / (25x^2 + 1):")
print(lagrange_f1)

print("\nPolinomio de Lagrange para f(x) = arctan(x):")
print(lagrange_f2)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Serie de Taylor
# 1. Función: f(x) = 1 / (25x^2 + 1), x0 = 0
x = sp.Symbol('x')
f1 = 1 / (25 * x**2 + 1)
x0_1 = 0

# Serie de Taylor para f1
n_terms = 4  # Número de términos de la serie
series_f1 = sp.series(f1, x, x0_1, n=n_terms).removeO()

print("Serie de Taylor para f(x) = 1 / (25x^2 + 1) alrededor de x0 = 0:")
print(series_f1)

# 2. Función: f(x) = arctan(x), x0 = 1
f2 = sp.atan(x)
x0_2 = 1

# Serie de Taylor para f2
series_f2 = sp.series(f2, x, x0_2, n=n_terms).removeO()

print("\nSerie de Taylor para f(x) = arctan(x) alrededor de x0 = 1:")
print(series_f2)

# Polinomio de Lagrange
# Datos para la interpolación de Lagrange
x_points = [0, 0.5, 1]
y_points_f1 = [f1.subs(x, xi) for xi in x_points]
y_points_f2 = [f2.subs(x, xi) for xi in x_points]

# Construcción del polinomio de Lagrange
def lagrange_polynomial(x_points, y_points):
    n = len(x_points)
    P = 0
    for i in range(n):
        L = 1
        for j in range(n):
            if i != j:
                L *= (x - x_points[j]) / (x_points[i] - x_points[j])
        P += y_points[i] * L
    return sp.simplify(P)

lagrange_f1 = lagrange_polynomial(x_points, y_points_f1)
lagrange_f2 = lagrange_polynomial(x_points, y_points_f2)

print("\nPolinomio de Lagrange para f(x) = 1 / (25x^2 + 1):")
print(lagrange_f1)

print("\nPolinomio de Lagrange para f(x) = arctan(x):")
print(lagrange_f2)

# Gráficas
x_vals = np.linspace(-1, 2, 500)

# Gráfica para f1
f1_func = sp.lambdify(x, f1, "numpy")
f1_taylor_func = sp.lambdify(x, series_f1, "numpy")
f1_lagrange_func = sp.lambdify(x, lagrange_f1, "numpy")

plt.figure(figsize=(12, 6))
plt.plot(x_vals, f1_func(x_vals), label="f(x) = 1 / (25x^2 + 1)", color="blue")
plt.plot(x_vals, f1_taylor_func(x_vals), label="Serie de Taylor", linestyle="--", color="orange")
plt.plot(x_vals, f1_lagrange_func(x_vals), label="Polinomio de Lagrange", linestyle="--", color="green")
plt.title("Aproximaciones para f(x) = 1 / (25x^2 + 1)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()

# Gráfica para f2
f2_func = sp.lambdify(x, f2, "numpy")
f2_taylor_func = sp.lambdify(x, series_f2, "numpy")
f2_lagrange_func = sp.lambdify(x, lagrange_f2, "numpy")

plt.figure(figsize=(12, 6))
plt.plot(x_vals, f2_func(x_vals), label="f(x) = arctan(x)", color="blue")
plt.plot(x_vals, f2_taylor_func(x_vals), label="Serie de Taylor", linestyle="--", color="orange")
plt.plot(x_vals, f2_lagrange_func(x_vals), label="Polinomio de Lagrange", linestyle="--", color="green")
plt.title("Aproximaciones para f(x) = arctan(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```
