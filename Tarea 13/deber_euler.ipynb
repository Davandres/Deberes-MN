**a) Utilice el m√©todo de Euler para aproximar las soluciones para cada uno de los siguientes problemas de valor inicial**

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method_a():
    # Definir las condiciones iniciales
    t0 = 1
    y0 = 1
    h = 0.1
    t_end = 2
    
    # Crear arrays para almacenar los valores de t y y
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    
    # M√©todo de Euler
    for i in range(1, len(t_values)):
        t = t_values[i-1]
        y = y_values[i-1]
        y_prime = y/t - (y/t)**2
        y_values[i] = y + h * y_prime
    
    # Graficar la soluci√≥n
    plt.plot(t_values, y_values, label='Aproximaci√≥n de Euler')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title('M√©todo de Euler para y\' = y/t - (y/t)^2')
    plt.legend()
    plt.grid()
    plt.show()

euler_method_a()
```

```python
def euler_method_b():
    # Definir las condiciones iniciales
    t0 = 1
    y0 = 0
    h = 0.2
    t_end = 3
    
    # Crear arrays para almacenar los valores de t y y
    t_values = np.arange(t0, t_end + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = y0
    
    # M√©todo de Euler
    for i in range(1, len(t_values)):
        t = t_values[i-1]
        y = y_values[i-1]
        y_prime = 1 + y/t + (y/t)**2
        y_values[i] = y + h * y_prime
    
    # Graficar la soluci√≥n
    plt.plot(t_values, y_values, label='Aproximaci√≥n de Euler')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title('M√©todo de Euler para y\' = 1 + y/t + (y/t)^2')
    plt.legend()
    plt.grid()
    plt.show()

euler_method_b()
```

aproximaciones del ejercicio 3.**

```python
import numpy as np
import matplotlib.pyplot as plt

def f_a(t, y):
    return y/t - (y/t)**2
def y_exacta_a(t):
    return t / (1 + np.log(t))

# Par√°metros
t0 = 1
y0 = 1
t_final = 2
h = 0.1

# Definir los puntos de t
t_values = np.arange(t0, t_final + h, h)
y_values = [y0]
y_exact_values_a = [y_exacta_a(t) for t in t_values]

# M√©todo de Euler
y = y0
for t in t_values[:-1]:
    y = y + h * f_a(t, y)
    y_values.append(y)

# Calcular el error real
errores_a = [abs(y_exact_values_a[i] - y_values[i]) for i in range(len(t_values))]


# Graficar la soluci√≥n exacta y la aproximada
plt.plot(t_values, y_exact_values_a, label="Soluci√≥n exacta", color='g')
plt.plot(t_values, y_values, label="Aproximaci√≥n de Euler", color='b', linestyle='--')
plt.xlabel('t')
plt.ylabel('y')
plt.title("Soluci√≥n exacta y aproximada - Problema a)")
plt.grid(True)
plt.legend()
plt.show()
```
