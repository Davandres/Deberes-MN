## Ejercicio 1: Sea ùëì(ùë•) = ‚àíùë•3 ‚àí cos ùë• y ùëù0 = ‚àí1. Use el m√©todo de Newton y de la Secante para encontrar ùëù2. ¬øSe podr√≠a usar ùëù0 = 0?

### Para el metodo de Newton:

```python
import numpy as np
import matplotlib.pyplot as plt

def newton_raphson(fx, dfx, xi, tolera, iteramax=2, vertabla=False, precision=4):
    '''
    M√©todo de Newton-Raphson para encontrar ra√≠ces.
    fx y dfx son funciones en forma de lambda (num√©ricas).
    xi es el punto inicial de b√∫squeda.
    tolera es la tolerancia permitida.
    iteramax es el m√°ximo n√∫mero de iteraciones permitidas.
    vertabla habilita o deshabilita la impresi√≥n de la tabla de iteraciones.
    '''
    itera = 0
    tramo = abs(2 * tolera)  # Inicialmente, aseguramos que el bucle comience
    if vertabla:
        print('M√©todo de Newton-Raphson')
        print('i', ['xi', 'fi', 'dfi', 'xnuevo', 'tramo'])
        np.set_printoptions(precision)
    
    while tramo >= tolera and itera < iteramax:
        fi = fx(xi)
        dfi = dfx(xi)
        xnuevo = xi - fi / dfi  # F√≥rmula de Newton-Raphson
        tramo = abs(xnuevo - xi)  # Error absoluto
        if vertabla:
            print(itera, np.array([xi, fi, dfi, xnuevo, tramo]))
        xi = xnuevo
        itera += 1

    if itera >= iteramax:
        xi = np.nan  # No converge
        print('Iteraciones: ', itera, 'No converge, se alcanz√≥ el m√°ximo de iteraciones')

    return xi

# INGRESO
fx = lambda x: -x**3 - np.cos(x)  # Usamos np.cos para la funci√≥n coseno
dfx = lambda x: -3 * (x**2) + np.sin(x)  # Usamos np.sin para la funci√≥n seno

x0 = -1  # Aproximaci√≥n inicial
tolera = 0.001  # Tolerancia

# PROCEDIMIENTO
respuesta = newton_raphson(fx, dfx, x0, tolera, vertabla=True)

# SALIDA
print('Ra√≠z en: ', respuesta)
```

### Con el metodo de la Secante:

```python
# M√©todo de secante
import numpy as np

def secante_raiz(fx,a,b,tolera, iteramax=20,
                 vertabla=True, precision=6):
    '''fx en forma num√©rica lambda
    Los valores de [a,b] son seleccionados
    desde la gr√°fica de la funci√≥n
    '''
    xi_1 = a
    xi = b
    itera = 0
    tramo = np.abs(xi-xi_1)
    if vertabla==True:
        print('m√©todo de la Secante')
        print('i','[ x[i-1], xi, x[i+1], f[i-1], fi ]','tramo')
        np.set_printoptions(precision)
    while not(tramo<tolera or itera>iteramax):
        fi_1 = fx(xi_1)
        fi = fx(xi)
        xi1 = xi-fi*(xi_1 - xi)/(fi_1-fi)
        tramo = np.abs(xi1-xi)
        if vertabla==True:
            print(itera,np.array([xi_1,xi,xi1,fi_1,fi]),tramo)
        xi_1 = xi
        xi = xi1
        itera = itera + 1
    if itera>=iteramax:
        xi = np.nan
        print('itera: ',itera,
              'No converge,se alcanz√≥ el m√°ximo de iteraciones')
    respuesta = xi
    
    return(respuesta)

# INGRESO
fx = lambda x: x**3 + 4*x**2 - 10
a = 1
b = 5
tolera = 0.001
c = (a+b)/2 # probar dos puntos en un mismo lado de f(x)

# PROCEDIMIENTO
respuesta = secante_raiz(fx,c,b,tolera,
                         vertabla=True,precision = 5)
# SALIDA
print('ra√≠z en: ', respuesta)
```

## EJERCICIO 6:La funci√≥n descrita por ùëì(ùë•) = ln(ùë•^2 + 1) ‚àí ‚ÑØ^0.4ùë•*cosùúãùë• tiene un n√∫mero infinito de ceros.
## a. Determine, dentro de 10^‚àí6, el √∫nico cero negativo.
## b. Determine, dentro de 10^‚àí6, los cuatro ceros positivos m√°s peque√±os.

## d. Use la parte c) para determinar, dentro de 10^‚àí6, el vigesimoquinto cero positivo m√°s peque√±o de ùëì.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Definici√≥n de la funci√≥n y su derivada
f = lambda x: np.log(x**2 + 1) - np.exp(0.4 * x) * np.cos(np.pi * x)
df = lambda x: (2 * x / (x**2 + 1)) - np.exp(0.4 * x) * (0.4 * np.cos(np.pi * x) - np.pi * np.sin(np.pi * x))

# Gr√°fica para identificar ceros
x_vals = np.linspace(-5, 30, 1000)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label="f(x)")
plt.axhline(0, color='red', linestyle='--')
plt.title("Gr√°fica de f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()
plt.show()

# a) Encontrar el √∫nico cero negativo
cero_negativo = newton(f, x0=-1, fprime=df, tol=1e-6)
print("√önico cero negativo:", cero_negativo)

# b) Encontrar los cuatro ceros positivos m√°s peque√±os
ceros_positivos = []
aproximaciones_iniciales = [0.5, 2.5, 4.5, 6.5]  # Aproximaciones iniciales basadas en la gr√°fica
for x0 in aproximaciones_iniciales:
    raiz = newton(f, x0=x0, fprime=df, tol=1e-6)
    ceros_positivos.append(raiz)

print("Cuatro ceros positivos m√°s peque√±os:", ceros_positivos)

# c) Aproximaci√≥n inicial para el en√©simo cero positivo
print("Aproximaci√≥n inicial para en√©simo cero positivo: Use la gr√°fica para identificar el patr√≥n.")

# d) Encontrar el vigesimoquinto cero positivo
aproximacion_25 = 2 * 25 - 0.5  # Patrones observados en la gr√°fica
cero_25 = newton(f, x0=aproximacion_25, fprime=df, tol=1e-6)
print("Vigesimoquinto cero positivo:", cero_25)
```
